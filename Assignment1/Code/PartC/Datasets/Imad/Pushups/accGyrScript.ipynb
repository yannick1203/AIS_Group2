```python
import pandas as pd
import numpy as np

acc = pd.read_csv("./Accelerometer.csv")
acc.drop(columns='Time (s)')
gyr = pd.read_csv("./Gyroscope.csv")
gyr.drop(columns='Time (s)')

activity = pd.DataFrame(columns= ["X (m/s^2)","Y (m/s^2)","Z (m/s^2)", "X (rad/s)", "Y (rad/s)", "Z (rad/s)"])

```

```python
activity['X (m/s^2)'] = acc['X (m/s^2)'].values
activity['Y (m/s^2)'] = acc['Y (m/s^2)'].values
activity['Z (m/s^2)'] = acc['Z (m/s^2)'].values
activity['X (rad/s)'] = gyr['X (rad/s)'].values
activity['Y (rad/s)'] = gyr['Y (rad/s)'].values
activity['Z (rad/s)'] = gyr['Z (rad/s)'].values

activity["Subject"] = 1
activity["Activity"] = "Pushups"
```

```python
activity.to_csv('imadPushups.csv', index=False)
```
