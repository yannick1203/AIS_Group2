```python
import pandas as pd

acc = pd.read_csv("Accelerometer.csv")
gyr = pd.read_csv("Gyroscope.csv")

activity = pd.DataFrame(columns= ["X (m/s^2)","Y (m/s^2)","Z (m/s^2)", "X (rad/s)", "Y (rad/s)", "Z (rad/s)"])

```

```python
activity
```

```python
activity['X (m/s^2)'] = acc['Acceleration x (m/s^2)'].values
activity['Y (m/s^2)'] = acc['Acceleration y (m/s^2)'].values
activity['Z (m/s^2)'] = acc['Acceleration z (m/s^2)'].values
activity['X (rad/s)'] = gyr['Gyroscope x (rad/s)'].values
activity['Y (rad/s)'] = gyr['Gyroscope y (rad/s)'].values
activity['Z (rad/s)'] = gyr['Gyroscope z (rad/s)'].values

activity["Subject"] = 3
activity["Activity"] = "Squads"
```

```python
activity
```

```python
activity.to_csv('yannickSquads.csv', index=False)
```
